데이터셋 분할

```python
import os
import random
import shutil

# 데이터 경로 설정
data_path = r'.\face_mask\face_mask'
train_path = r'.\face_mask\train'
val_path = r'.\face_mask\val'
test_path = r'.\face_mask\test'

# 데이터 파일 리스트
all_files = os.listdir(data_path)
random.shuffle(all_files)

# 데이터 분할
train_ratio = 0.6
val_ratio = 0.2
test_ratio = 0.2

train_files = all_files[:int(train_ratio * len(all_files))]
val_files = all_files[int(train_ratio * len(all_files)):int((train_ratio + val_ratio) * len(all_files))]
test_files = all_files[int((train_ratio + val_ratio) * len(all_files)):]

# 디렉토리 생성 및 파일 복사
for file_set, dest_path in zip([train_files, val_files, test_files], [train_path, val_path, test_path]):
    os.makedirs(dest_path, exist_ok=True)
    for file_name in file_set:
        shutil.copy(os.path.join(data_path, file_name), dest_path)
```
